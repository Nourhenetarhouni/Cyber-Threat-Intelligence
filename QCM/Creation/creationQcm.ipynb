```python
import csv
import re

# Function to read the TSV file and parse its content
def parse_generated_tsv(input_file, output_file):
    with open(input_file, 'r', newline='', encoding='utf-8') as tsvfile:
        reader = csv.reader(tsvfile, delimiter='\t')
        header = next(reader)

        # Assume the content is in the first row and first column
        for row in reader:
            generated_content = row[0]
            break

    lines = generated_content.split('\n')
    questions = []
    current_question = None
    answer_key = {}

    question_pattern = re.compile(r"Question \d+:")
    option_pattern = re.compile(r"\((A|B|C|D)\)")

    for line in lines:
        if question_pattern.search(line):
            if current_question:
                questions.append(current_question)
            current_question = {"Question": question_pattern.split(line)[1].strip()}
        elif option_pattern.search(line):
            options = option_pattern.split(line)
            if current_question:
                current_question["Question"] += " " + options[0].strip()
            for i in range(1, len(options) - 1, 2):
                option_label = options[i]
                option_text = options[i + 1].strip()
                current_question[f"Option {option_label}"] = option_text
        elif line.startswith("1.") or line.startswith("2.") or line.startswith("3.") or line.startswith("4.") or line.startswith("5."):
            parts = line.split(". ")
            question_number = parts[0].strip()
            correct_answer = parts[1].strip()
            answer_key[question_number] = correct_answer

    if current_question:
        questions.append(current_question)

    for i, question in enumerate(questions):
        question_number = str(i + 1)
        if question_number in answer_key:
            question["Correct Answer"] = answer_key[question_number]

    # Writing to TSV file with new structure
    with open(output_file, 'w', newline='', encoding='utf-8') as tsvfile:
        writer = csv.DictWriter(tsvfile, fieldnames=["Question", "Option A", "Option B", "Option C", "Option D", "Correct Answer"], delimiter='\t')
        writer.writeheader()
        for question in questions:
            writer.writerow(question)

# Specify the input and output file paths
input_file = '/content/questions.tsv'
output_file = '/content/parsed_questions.tsv'

# Parse the input TSV file and write to the output TSV file
parse_generated_tsv(input_file, output_file)

print(f'Parsed TSV file has been saved to {output_file}')
```

```python
import os
import re

# Spécifiez le répertoire contenant les fichiers et le fichier de sortie
input_dir = '/content'  # Remplacez par votre répertoire si différent
output_file = '/content/questions.txt'

# Fonction pour extraire les informations du premier format
def parse_question_file_format1(contents):
    questions = []
    # Utiliser des chaînes brutes pour éviter les problèmes avec les caractères spéciaux
    pattern = re.compile(
        r'\*\*Question \d+\*\*\s+(.*?)\n'
        r'(A\. .+\nB\. .+\nC\. .+\nD\. .+\n)'
        r'\*\*Correct Answer:\*\* ([A-D])\n'
        r'(?:\*\*Explanation:\*\*\s+(.*))?',
        re.DOTALL
    )
    matches = pattern.finditer(contents)
    for match in matches:
        question_text, options_text, correct_answer, explanation = match.groups()

        # Séparer les options
        options = re.findall(r'([A-D])\. (.+)', options_text)

        # Formater la question
        formatted_question = (
            '**Question**\n{}\n{}\n**Correct Answer:** {}\n{}'.format(
                question_text.strip(),
                '\n'.join(['Option {}: {}'.format(opt[0], opt[1]) for opt in options]),
                correct_answer.strip(),
                '**Explanation:** {}'.format(explanation.strip()) if explanation else ''
            )
        )
        questions.append(formatted_question)
    return questions

# Fonction pour extraire les informations du deuxième format
def parse_question_file_format2(contents):
    questions = []
    pattern = re.compile(
        r'\| \*\*MITRE ATT&CK [^\*]+\*\*\n'
        r'\|([^|]+?)\n'
        r'\| Option A: ([^|]+?) \| Option B: ([^|]+?) \| Option C: ([^|]+?) \| Option D: ([^|]+?) \| Correct Answer: ([A-D]) \|'
        r'(?: Explanation: ([^|]+?))?\|',
        re.DOTALL
    )
    matches = pattern.finditer(contents)
    for match in matches:
        question_text, option_a, option_b, option_c, option_d, correct_answer, explanation = match.groups()

        # Formater la question
        formatted_question = (
            '**Question**\n{}\nOption A: {}\nOption B: {}\nOption C: {}\nOption D: {}\n**Correct Answer:** {}\n{}'.format(
                question_text.strip(),
                option_a.strip(),
                option_b.strip(),
                option_c.strip(),
                option_d.strip(),
                correct_answer.strip(),
                '**Explanation:** {}'.format(explanation.strip()) if explanation else ''
            )
        )
        questions.append(formatted_question)
    return questions

# Ouvrez le fichier de sortie en mode écriture
with open(output_file, 'w') as outfile:
    # Liste les fichiers dans le répertoire spécifié
    files = [f for f in os.listdir(input_dir) if f.endswith('.txt') and f.startswith('question')]
    if not files:
        print("Aucun fichier de question trouvé.")
    else:
        print(f"Traitement de {len(files)} fichiers...")
        # Parcourez tous les fichiers dans le répertoire
        for filename in sorted(files):  # Trier les fichiers pour un traitement ordonné
            filepath = os.path.join(input_dir, filename)
            print(f"Lecture du fichier : {filename}")
            with open(filepath, 'r') as infile:
                contents = infile.read()
                # Analysez le contenu et obtenez les questions pour les deux formats
                questions = parse_question_file_format1(contents) + parse_question_file_format2(contents)
                # Écrivez les questions dans le fichier de sortie
                for question in questions:
                    outfile.write(question + '\n\n')

print(f"Le fichier intégré a été créé : {output_file}")
```

```python
import re
import csv

def parse_qcm_file(file_path):
    data = []
    question, option_a, option_b, option_c, option_d, correct_answer, explanation = "", "", "", "", "", "", ""

    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

        for line in lines:
            line = line.strip()
            if not line:
                continue

            # Print each line for debugging
            print(f"Processing line: {line}")

            # Check for each section
            if line.startswith("**Question"):
                # If we already have a question, save it before starting a new one
                if question:
                    data.append([question, option_a, option_b, option_c, option_d, correct_answer, explanation])
                    option_a, option_b, option_c, option_d, correct_answer, explanation = "", "", "", "", "", ""  # Reset variables

                question = line.split("**Question", 1)[1].strip()  # Capture the question text

            elif "**Option A:**" in line:
                option_a = line.split("**Option A:**", 1)[1].strip()
                print(f"Found Option A: {option_a}")  # Debugging

            elif "**Option B:**" in line:
                option_b = line.split("**Option B:**", 1)[1].strip()
                print(f"Found Option B: {option_b}")  # Debugging

            elif "**Option C:**" in line:
                option_c = line.split("**Option C:**", 1)[1].strip()
                print(f"Found Option C: {option_c}")  # Debugging

            elif "**Option D:**" in line:
                option_d = line.split("**Option D:**", 1)[1].strip()
                print(f"Found Option D: {option_d}")  # Debugging

            elif "**Correct Answer:**" in line:
                correct_answer = line.split("**Correct Answer:**", 1)[1].strip()
                print(f"Found Correct Answer: {correct_answer}")  # Debugging

            elif "**Explanation:**" in line:
                explanation = line.split("**Explanation:**", 1)[1].strip()
                print(f"Found Explanation: {explanation}")  # Debugging

        # Add the last question to the data list
        if question:
            data.append([question, option_a, option_b, option_c, option_d, correct_answer, explanation])

    return data

def write_to_tsv(data, output_file):
    with open(output_file, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f, delimiter='\t')
        writer.writerow(['Question', 'Option A', 'Option B', 'Option C', 'Option D', 'Correct Answer', 'Explanation'])
        writer.writerows(data)

# Specify the input and output file paths
input_file = 'questions.txt'
output_file = 'qcm_output.tsv'

# Parse the file and write to TSV
qcm_data = parse_qcm_file(input_file)
write_to_tsv(qcm_data, output_file)
```

```python
import re

def display_statistics(input_file):
    # Counters for the statistics
    question_count = 0
    option_count = 0
    correct_answer_count = 0
    explanation_count = 0

    # Regular expressions to match questions, options, correct answers, and explanations
    question_pattern = re.compile(r'\*\*Question.*?\*\*(.*?)\n', re.DOTALL)
    option_pattern = re.compile(r'\*\*Option A:\*\*(.*?)\n\*\*Option B:\*\*(.*?)\n\*\*Option C:\*\*(.*?)\n\*\*Option D:\*\*(.*?)\n', re.DOTALL)
    correct_answer_pattern = re.compile(r'\*\*Correct Answer:\*\*(.*?)\n', re.DOTALL)
    explanation_pattern = re.compile(r'\*\*Explanation:\*\*(.*?)\n', re.DOTALL)

    with open(input_file, 'r', encoding='utf-8') as file:
        content = file.read()

        question_count = len(question_pattern.findall(content))
        option_count = len(option_pattern.findall(content))
        correct_answer_count = len(correct_answer_pattern.findall(content))
        explanation_count = len(explanation_pattern.findall(content))

    # Assuming valid entries are questions that have all necessary components
    valid_entries_count = min(question_count, option_count, correct_answer_count, explanation_count)

    # Print the statistics
    print(f"Found {question_count} questions")
    print(f"Found {option_count} options")
    print(f"Found {correct_answer_count} correct answers")
    print(f"Found {explanation_count} explanations")
    print(f"Valid entries count: {valid_entries_count}")

if __name__ == "__main__":
    input_file = 'questions.txt'  # Replace with your actual file path
    display_statistics(input_file)
```
